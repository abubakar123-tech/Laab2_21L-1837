```python
from google.colab import files
import numpy as np
import matplotlib.pyplot as plt

uploaded = files.upload()
with open('genome.txt', 'r') as file:
    genome_sequence = file.read().strip()

genome_list = list(genome_sequence)
genome_length = len(genome_list)

t = np.linspace(0, 4 * np.pi, genome_length)
x = np.cos(t)
y = np.sin(t)
z = np.linspace(0, 5, genome_length)
coordinates = np.column_stack((x, y, z))

color_map = {'A': 'red', 'T': 'blue', 'C': 'green', 'G': 'orange'}
colors = [color_map.get(nuc, 'black') for nuc in genome_list]

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(coordinates[:, 0], coordinates[:, 1], coordinates[:, 2], c=colors, s=30)
ax.set_title('3D Helix Structure of Genome Sequence')
ax.set_xlabel('X-axis (cos(t))')
ax.set_ylabel('Y-axis (sin(t))')
ax.set_zlabel('Z-axis (linear)')
plt.show()
```

```python
from google.colab import files
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

uploaded = files.upload()
filename = next(iter(uploaded))

img = Image.open(filename)
img_array = np.array(img)

# Step 2: Plot the original image
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.imshow(img_array)
plt.title('Original Image')
plt.axis('off')

# Step 3: Rotate the image 90 degrees
rotated_img = np.rot90(img_array)
plt.subplot(2, 2, 2)
plt.imshow(rotated_img)
plt.title('Rotated Image (90)')
plt.axis('off')

# Step 4: Flip the image horizontally
flipped_img = np.fliplr(img_array)
plt.subplot(2, 2, 3)
plt.imshow(flipped_img)
plt.title('Flipped Image (Left-Right)')
plt.axis('off')

# Step 5: Convert to grayscale using the formula
gray_img = np.dot(img_array[..., :3], [0.299, 0.587, 0.114])
plt.subplot(2, 2, 4)
plt.imshow(gray_img, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# Show all plots
plt.tight_layout()
plt.show()
```

```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt

iris = load_iris()
X, Y = iris.data, iris.target

print("Mean:", np.mean(X, axis=0))
print("Median:", np.median(X, axis=0))
print("Std Dev:", np.std(X, axis=0))
print("Min:", np.min(X, axis=0))
print("Max:", np.max(X, axis=0))

sepal = X[:, :2]

plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.scatter(sepal[:, 0], sepal[:, 1], c=Y, cmap='viridis')
plt.title("Sepal Length vs Width")
plt.xlabel("Length"); plt.ylabel("Width")

plt.subplot(132)
plt.hist(sepal[:, 0], bins=15, color='skyblue', edgecolor='black')
plt.title("Sepal Length Histogram")

plt.subplot(133)
plt.plot(X[:, 2], X[:, 3], 'r.')
plt.title("Petal Length vs Width")
plt.xlabel("Length"); plt.ylabel("Width")

plt.tight_layout()
plt.show()
```
